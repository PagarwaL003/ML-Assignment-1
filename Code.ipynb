```python
# import dependenices
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import boston housing data set
data = pd.read_csv("Data.csv")
```

```python
data
```

```python
# check for null values
data = data.dropna()
data.isnull().sum()
```

```python
# data
data
```

```python
X = data.drop(columns = ["medv"])
X
```

```python
y = data["medv"]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
lr.fit(X_train, y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
y_predict_train = lr.predict(X_train)
y_predict_train
```

```python
plt.scatter(y_train, y_predict_train)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, y_predict_train)
```

```python
y_predict_test = lr.predict(X_test)
```

```python
plt.scatter(y_test, y_predict_test)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
score = r2_score(y_test, y_predict_test)
print("Score =", score)
```

```python
# calculating error lowest as possible
from sklearn.metrics import mean_squared_error, mean_absolute_error
sqE = mean_squared_error(y_test, y_predict_test)
absE = mean_absolute_error(y_test,y_predict_test)
print("Mean Square Error = ", sqE)
print("Mean Absolute Error = ", absE)
```
